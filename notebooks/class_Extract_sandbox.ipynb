```python
import os
import pandas as pd
```

```python
class Extract:
    def __init__(self, url, parquet_file_name="output.parquet", csv_file_name="output.csv"):
        self.url = url
        self.parquet_file_name = parquet_file_name
        self.csv_file_name = csv_file_name
        self.__df = None

    def download_parquet(self):
        os.system(f"wget {self.url} -O {self.parquet_file_name}")

    def convert_parquet_to_csv(self):
        df = pd.read_parquet(self.parquet_file_name)
        df.to_csv(self.csv_file_name)
        self.__df = df

    def n_rows(self):
        n_rows = self.__df.shape[0]
        return n_rows
```

```python
URL = "https://d37ci6vzurychx.cloudfront.net/trip-data/yellow_tripdata_2021-01.parquet"
extract = Extract(url=URL)
```

```python
extract.download_parquet()
```

```python
extract.convert_parquet_to_csv()
```

```python
extract.n_rows()
```

```python
from dataclasses import dataclass
import os
import pandas as pd

@dataclass
class Extract:
    url: str
    parquet_file_name: str = "output.parquet"
    csv_file_name: str = "output.csv"
    __df = None
    n_rows = None

    def download_parquet(self):
        os.system(f"wget {self.url} -O {self.parquet_file_name}")

    def get_pandas_df(self):
        self.__df = pd.read_parquet(self.parquet_file_name)
        self.n_rows = self.__df.shape[0]
        return self.__df

    def convert_parquet_to_csv(self):
        self.__df.to_csv(self.csv_file_name)
```

```python
extract = Extract("https://d37ci6vzurychx.cloudfront.net/trip-data/yellow_tripdata_2021-01.parquet")
```
